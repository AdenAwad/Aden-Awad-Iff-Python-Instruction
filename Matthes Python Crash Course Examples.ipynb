```python
# ex. 1: lists

bicycles = ['trek', 'cannondale', 'redline', 'specialized']
print(bicycles) 
# list is printed with brackets, quotes, and commas

print(bicycles[0])
# index of an entry in square brackets after name of list accesses that entry. index positions start at 0, not 1. 
# -1 indexes the last element, -2 the second to last, etc...

print(bicycles[0].title())
# string method used on list entry

message =  f"my first bicycle was a {bicycles[3].title()}"
print(message)
```

```python
# ex. 2: adding and removing elements

motorcycles = ['honda','yamaha', 'suzuki']
print(motorcycles)

motorcycles[0] = 'ducati'
print(motorcycles)

motorcycles.append('harley')
print(motorcycles)

del motorcycles[3]
print(motorcycles)

popped_motorcycle = motorcycles.pop()
print(motorcycles)
print(popped_motorcycle)
# pop() method removes last element by default, but index pay be specified in ()
# remove() method works the same way, but specifies value not index  in (). deletes only the first instance of a value in a list
```

```python
# ex. 1: for loop

hobbits = ['frodo','sam','merry', 'pippin']
for hobbit in hobbits:
    print(hobbit)

for hobbit in hobbits:
    print(f"\nI don't think he knows about second breakfast, {hobbit}.")

# colon tells python to interpret next line as start of loop
# any code after the loop that is not indented is not controlled by the loop, so can show that a loop has finished executing and summarize results
```

```python
# ex. 2: numerical lists

for value in range(1, 5):
    print(value)
# doesn't return last value by 'start at 0' convention. list() function formats output as list.

evens = list(range(0,11,2))
print(evens)
# third argument in range() specifies step size

squares = []
for value in range(0,11):
    square = value ** 2
    squares.append(square)
print(squares)

# alternatively:
squares1 = []
for value in range(0,11):
    squares1.append(value ** 2)
print(squares1)

# alternatively using list comprehensions:
squares2 = [value**2 for value in range(0,11)]
print(squares2)
# min(), max(), and sum() functions do what's expected

# list[a:b] gives a 'slice' of list from element a to element b. 
# list[a:] slices from element a to the last element; list[:b] slices from the first element to element b
# list[:] (removing both index arguments) copies the list
```

```python
# ex. 3: tuples

# tuples are immutable lists defined using commas separating elements, conventionally w/ () instead of [].
# a tuple with 1 element must n be written (n,)

T = (1,10)
for t in T:
    print(t)

T = (1,11)
print('\nnew T:')
for t in T:
    print(t)
```

```python
# ex. 1: if statements

cars = ['prius', 'camry', 'accord', 'cybertruck']
for car in cars:
    if car == 'cybertruck':
        print(f'{car.title()} (eww)')
    else:
        print(car.title())
# == is the equality operator, returning true if the two values match. 
# testing equality of strings is case sensitive. to bypass this, use str.lower() == 'xyz'
# != similarly functions as the inequality operator

# numerical comparison operators >, <, >=, <= follow same logic

order = ['burger', 'salad', 'coffee', 'sweet potato fries']
bad = 'sweet potato fries'

if bad in order:
    order.remove(bad)
    print(f"let's get {order} instead.")
```

```python
# ex. 2: if-else and if-elif-else

age = 21
if age - 4 > 18:
    print("you're old enough to vote")
elif 18 < age <= 22:
    print("first time, huh?")
else:
    print("you're too young")
    
```
