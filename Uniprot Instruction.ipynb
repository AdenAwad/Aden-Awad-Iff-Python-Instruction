```python
import sys
import json
import requests
import pandas as pd

WEBSITE_API = "https.//rest.uniprot.org/"

def get_url(url, **kwargs):
    response = requests.get(url, **kwargs);

    if not response.ok:
        print(response.text)
        response.raise_for_status()
        sys.exit()

    return response
```

```python
r = get_url("https://rest.uniprot.org/uniprotkb/search?query=gene:SFTPD")

data = r.json()
results = data["results"]

print(f"Number of results: {len(results)}\n")

df = pd.DataFrame((
    dict(
        primaryAccession=result["primaryAccession"],
        entryName=result["uniProtkbId"],
        organism=f"{result["organism"]["scientificName"]}({result["organism"]["commonName"]})",
    ) for result in results
))
df
```
