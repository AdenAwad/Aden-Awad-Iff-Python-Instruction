```python
# Individual Synthesis Example: Domain coloring plot of the Riemann zeta function
# I synthesized skills I learned from the examples to create my own plot.

# zeta(s) is computed from the Dirichlet eta function:
#     eta(s) = sum_{n=1}^N (-1)^(n-1) / n^s
#     zeta(s) = eta(s) / (1 - 2^(1 - s))
# The alternating eta series converges for Re(s) > 0 and, with enough terms, behaves well across the strip we plot. We use Euler (averaged partial-sum) acceleration so a modest N gives a clean picture, including Re(s) < 0 via the same eta/zeta relation.

# Plotting uses Quick Start Guide techniques:
#   - object-oriented style: fig, ax = plt.subplots()      (Examples 1, 3.1)
#   - np.meshgrid grid + pcolormesh + colorbar             (Example 14)
#   - contour overlay                                      (Example 14)
#   - set_xlabel / set_ylabel / set_title                  (Example 3.1)
#   - ax.text + ax.plot annotation                         (Examples 7, 8)

import matplotlib.pyplot as plt
import numpy as np

# ---------------------------------------------------------------
# Vectorized zeta over a complex array, using an Euler-accelerated
# alternating series for eta(s). Works for an array S of any shape.
# ---------------------------------------------------------------
def zeta(S, N=40):
    S = np.asarray(S, dtype=complex)

    # Euler summation weights for the alternating series sum (-1)^(n-1) a_n.
    # w_n are the averaged-partial-sum (binomial) weights; this turns a slowly/
    # non-convergent alternating tail into an accurate value.
    # Reference identity: eta(s) = sum_n e_n * (-1)^(n-1) / n^s  with the
    # weights below (van Wijngaarden / Euler transform, truncated at N).
    n = np.arange(1, N + 1)

    # binomial-tail weights: e_k = 2^-N * sum_{j>=k} C(N, j)
    from math import comb
    binom = np.array([comb(N, j) for j in range(N + 1)], dtype=float)
    tail = np.cumsum(binom[::-1])[::-1]          # tail[k] = sum_{j>=k} C(N,j)
    e = tail[1:] / (2.0 ** N)                    # length N, weights for n=1..N

    # eta(s) = sum_{n=1}^N e_n * (-1)^(n-1) * n^{-s}
    # broadcast: S has shape (...), n has shape (N,)
    n_pow = n[:, None] ** (-S.ravel()[None, :])  # (N, M)
    signs = ((-1.0) ** (n - 1))[:, None]         # (N, 1)
    eta = np.sum(e[:, None] * signs * n_pow, axis=0)  # (M,)

    eta = eta.reshape(S.shape)
    return eta / (1.0 - 2.0 ** (1.0 - S))

# Build the complex grid with meshgrid 
X, Y = np.meshgrid(np.linspace(-10, 10, 360),
                   np.linspace(-25, 25, 450))
S = X + 1j * Y

W = zeta(S)

phase = np.angle(W)          # in [-pi, pi]  -> the colored field
magnitude = np.abs(W)        # used for the contour overlay

# --- Plot: pcolormesh for phase + colorbar, like Cell 22 ---
fig, ax = plt.subplots(figsize=(7, 8), layout='constrained')

pc = ax.pcolormesh(X, Y, phase, cmap='hsv', vmin=-np.pi, vmax=np.pi)
cbar = fig.colorbar(pc, ax=ax)
cbar.set_label('phase  arg ζ(s)  [rad]')
cbar.set_ticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
cbar.set_ticklabels(['−π', '−π/2', '0', 'π/2', 'π'])

# Contour overlay of magnitude 
co = ax.contour(X, Y, magnitude,
                levels=[0.25, 0.5, 1, 2, 4, 8],
                colors='black', linewidths=0.5, alpha=0.5)
ax.clabel(co, inline=True, fontsize=6)

# mark the critical line and the pole (text/plot as in Examples 7, 8)
ax.plot([0.5, 0.5], [-25, 25], color='white', linewidth=1, linestyle='--')
ax.plot(1, 0, 'wo')
ax.text(0.7, 23, 'critical line', color='white', fontsize=8)

# labels and title in the OO style (Example 3.1)
ax.set_xlabel('Re(s)')
ax.set_ylabel('Im(s)')
ax.set_title('Individual Synthesis Example: Domain Coloring Plot of the Riemann Zeta Function ζ(s)', color='blue')

fig.savefig('zeta_pure_numpy.png', dpi=150)
```
